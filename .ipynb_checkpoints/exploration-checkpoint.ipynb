### Our data structure

> Our final dataset is composed of 104505 rows and 20 columns, there is no missing values and most of the columns provide informations about lenders financial situation and profile at the moment of the data recording. 

### Main feature of interest

> The main feature of interest in the dataset is  **LP_CollectionFees**, it represents how much it costs **Prosper Loan** to collect payments from borrowers. Higher is the collection fees, lower is the account value in termn of profitability to the company. Therefore, we will look at what are the characteristics of a bad borrower i.e. a borrower that is less/not profitable for the company.

### Features supporting our investigation

> It is not clear yet what may be all the features that could help support our investigation. We start be using common sense and intuition to explore a set features and eventually add some more if necessary. So, the features we will be starting with are the following:  **'EmploymentStatus'**,  **'Occupation'**, **'BorrowerState'**, **'CreditScore'**, **'DebtToIncomeRatio'**, because we are hypothesizing that those variables could be determinant on the ability of an applicant to qualify for a loan, get a competitive rate and be able to repay it back properly. CreditScore is not a feature that currently exist in our dataset, however we will be able to explore it by creating a new variable of the same name that will be equal to the average of CreditScoreRangeLower and CreditScoreRangeUpper. Before going any further, let's create it.  

```python
df['CreditScore'] = (df['CreditScoreRangeLower'] + df['CreditScoreRangeUpper'])/2
```

### CreditScore

A credit score of 700 and above is considered a good one.

```python
df1.CreditScore.describe()
```

```python
# What are the outliers boundaries?
q1, q3= np.percentile(df1.CreditScore,[25,75])
iqr = q3 - q1
lower_bound = q1 - (1.5 * q1)
upper_bound = q3 + (1.5 * q3)
print("Outlier Lower Boundary: {}\nOutlier Upper Boundary: {}".format(lower_bound, upper_bound))
```

```python
# We query the above average D/I (Debt-to-income) ratio
bad_credit = sum(df1.query("CreditScore < 700")['CreditScore'].value_counts())

# We query the below average D/I (Debt-to-income) ratio
good_credit = sum(df1.query("CreditScore >= 700")['CreditScore'].value_counts())

# we draw a pie chart
sizes = [bad_credit, good_credit]
labels = ['below average credit score', 'Above average credit score']
explode = (0, 0.1)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode)
plt.tight_layout()
plt.title('Credit Score distribution')
plt.show()
```

As we can see 71.7% of borrowers with high collection fees also have below average credit score. What about debt-to-income ratio?

### DebtToIncomeRatio

```python
# What's the statistical distribution?
df1.DebtToIncomeRatio.describe()
```

```python
# What are the outliers boundaries?
q1, q3= np.percentile(df1.DebtToIncomeRatio,[25,75])
iqr = q3 - q1
lower_bound = q1 - (1.5 * q1)
upper_bound = q3 + (1.5 * q3)
print("Outlier Lower Boundary: {}\nOutlier Upper Boundary: {}".format(lower_bound, upper_bound))
```

A debt-to-income ratio of 36% is considered reasonable to still qualify for lending products with most financial services companies. So let's look at the percentage of people with collection fees that have a below average debt-to-income ratio.

```python
x = 100*df1.query("DebtToIncomeRatio > 0.36")['DebtToIncomeRatio'].count()/df1.shape[0]
x = round(x, 0)
print("Only {}% of the borrowers having a high collection fees also have a below average debt-to-income ratio.".format(x) )
```

```python
# We query the above average D/I (Debt-to-income) ratio
good_ratio = sum(df1.query("DebtToIncomeRatio <= 0.36")['DebtToIncomeRatio'].value_counts())

# We query the below average D/I (Debt-to-income) ratio
bad_ratio = sum(df1.query("DebtToIncomeRatio > 0.36")['DebtToIncomeRatio'].value_counts())

# we draw a ppie chart
sizes = [good_ratio, bad_ratio]
labels = ['Above Average D/I ratio', 'Below Average D/I ratio']
explode = (0, 0.1)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode)
plt.tight_layout()
plt.title('Debt-to-income ratio distribution')
plt.show()
```

82.6% of the borrowers with high collection fees have an above average debt-to-income ratio.

## Bivariate Exploration

### LP_CollectionFees - Occupation

```python
occupation = df1.query("Occupation == ('Professional','Clerical','Teacher','Administrative Assistant','Sales - Commission')")['Occupation']
collection_fees = df1.query("Occupation == ('Professional','Clerical','Teacher','Administrative Assistant','Sales - Commission')")['LP_CollectionFees']
plt.figure(figsize=(16,6))
sb.violinplot(data = df, x = occupation, y = collection_fees)
plt.title('Relationship between occupation and collection fees')
plt.xticks(rotation=0);
```

The majority of **occupations** have their collection fees around zero and there is significantly fewer observations while we move below zero. The occupation where we observe the highest collection fees  are respectively: Teacher, Prefessional, Clerical, Sales-commission, and finally Administrative Assistant. 

### LP_CollectionFees - EmploymentStatus

```python
employmentStatus = df1.query("EmploymentStatus == ('Full-time','Employed','Self-employed')")['EmploymentStatus']
plt.figure(figsize=(16,6))
sb.violinplot(data = df, x = employmentStatus, y = collection_fees)
plt.title('Relationship between employment status and collection fees')
plt.xticks(rotation=0);
```

We can make approximately the same observation as above in terms of the distribution of data with people working full-time prsenting the highest collection fees, followed by people that are employed and finally those that are self-employed. 

###  LP_CollectionFees - BorrowerState

```python
borrowerState = df1.query("BorrowerState == ('CA','IL','GA','NY','FL')")['BorrowerState']
plt.figure(figsize=(16,6))
sb.violinplot(data = df, x = borrowerState, y = collection_fees)
plt.title('Relationship between borrower state and collection fees')
plt.xticks(rotation=0);
```

Here we notice that Georgia has the highest collection fees, followed by California, Illinois, New York and last Florida,

### LP_CollectionFees - CreditScore

```python
creditScore = df1.CreditScore
collectionFees = df1.LP_CollectionFees
plt.figure(figsize=(16,6))
plt.title('Relationship between credit score and collection fees')
sb.regplot(x=creditScore, y=collectionFees);
```

It seems like people with the best credit score are the one with the highest collection fees. 

### LP_CollectionFees - DebtToIncomeRatio

```python
debtToIncome = df1.DebtToIncomeRatio
plt.figure(figsize=(16,6))
sb.regplot(x=debtToIncome, y=collectionFees)
plt.title('Relationship between debt-to-income ratio and collection fees');
```

As with credit score, here we can see that people with the best debt-to-income ratio are the one with the highest collection fees. 

### DebtToIncome - CreditScore

```python
# Group credit score data in bins
bins = list(np.arange(0,1100,350))
grades = ['very bad', 'bad', 'good']
df1['CreditGrade'] = pd.cut(df1.CreditScore, bins, labels=grades);
```

```python
debtToIncome = df1.DebtToIncomeRatio
creditgrade = df1.CreditGrade
plt.figure(figsize=(16,6))
sb.violinplot(data = df1, x = creditgrade, y = debtToIncome)
plt.title('Relationship between credit grade and debt-to-income ratio')
plt.xticks(rotation=0);
```

People with a very bad credit score have for the majority a debt-to-income ratio around 0 which make sense as they could not qualify for more loans, people with average credit (>350 and <700) and those with a good one (>700) have for the majority a debt-to-income ratio around 0 and 1 and some very few observations having high to very high ratio. 



### Bivariate exploration conclusion
> Globally we noticed that the most extreme values in terms of collection fees have been noticed with borrowers that have **"teacher"** as occupation, people who are working **"full-time"**, leaving in **"Georgia"** state, however the later is not something that explain their behaviour and therefore is not a significant point to raise.  Something very particular, however, is that people with the best credit score and debt-to-income ratio are the one with the highest collection fees. It is also worthy to mention that it doesn't seem like those observed characteristics follow a strong pattern for bad debtors. In fact, so far what we can notice is that those bad debtors seems to be outliers.

## Multivariate Exploration

### Occupation - LP_CollectionFees - EmploymentStatus

```python
occupation = df1.query("Occupation == ('Professional','Clerical','Teacher','Administrative Assistant','Sales - Commission')")['Occupation']
collection_fees = df1.query("Occupation == ('Professional','Clerical','Teacher','Administrative Assistant','Sales - Commission')")['LP_CollectionFees']
employmentStatus = df1.query("EmploymentStatus == ('Full-time','Employed','Self-employed')")

# We assign a shape to each employment status
status_shape = [['Employed','D'], ['Full-time', 's'], ['Self-employed', 'v']]

plt.figure(figsize=(20,8))

for status, shape in status_shape:    
    plt.scatter(data = employmentStatus, x = occupation, y = collection_fees, marker = shape)

plt.legend(['Employed', 'Full-time', 'Self-employed'], title="EmploymentStatus")
plt.xlabel('Occupation')
plt.ylabel('LP_CollectionFees')
plt.title('Relationship between occupation, emploment status, and collection fees');
```

It doesn't look like **EMployment status** data is bringing any more significant insight than what provided by **Collection fees** and **Occupation**. 

### Occupation - LP_CollectionFees - CreditScore

```python
collectionFees = df1.LP_CollectionFees

creditgrade_shape = [['very bad','D'], ['bad', 's'], ['good', 'v']]

plt.figure(figsize=(20,8))

for status, shape in creditgrade_shape:
    plt.scatter(data = creditgrade, x = occupation, y = collection_fees, marker = shape)

plt.legend(['very bad', 'bad', 'good'], title="CreditScore")
plt.xlabel('Occupation')
plt.ylabel('LP_CollectionFees')
plt.title('Relationship between occupation, credit score, and collection fees');
```

We get to the same conclusion here than with the previous visualization. 

### Conclusion
> From the two last visualizations, the take away is is that the distribution of the observation doesn't change whether we look at different third variable like **EmploymentStatus** and **CreditScore**, with **Occupation** and **LP_CollectionFees** on the axes. We noticed that overall, borrowers do have below average credit score prior to qualify for a loan. Therefore what we conclude is that bad borrower behaviours seem to be related to a historic of wrong management of bill/debt payment even with sufficiently good financial standing to afford those pyment. 
We also have so isolated cases of borrowers with above average credit score who fell in the category of high collection fees, the justification is that they have started having this bad behaviour after they got the loan but this still confirm that those collection fees are by the fact that the company has hard time getting the money back from the borrowers because they are just not making their monthly payments even though they could definitely afford them.
